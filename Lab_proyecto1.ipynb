# Importación de librerías y definición de variables globales

```python
# Solución de la actividad #2

# Importación de librerías
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Se definen parámetros globales necesarios para generar las ondas
# Definición de la frecuencia y aplitud para las señales
frequency = 0.007
amplitude = 1.0

# Definición de la tasa de muestreo y la duración de la onda
samplingRate = 10
duration = 800

# Se genera el time array
t = np.arange(0, duration, 1/samplingRate)
```

# Definición de las función generadora de señal

En esta parte se genera una función que recibe los parámetros de: frecuencia, amplitud, el tipo de onda deseada por el usuario y el array de tiempo y a partir de estas entradas va a generar una onda de salida. Para generar los tipos de onda solicitados se utilizó como base las librerías numpy y scipy.



```python
# Función que genera la onda limpia de acuerdo al valor de waveType y al valor de
# las variables globales de frecuencia, amplitud y array de tiempo.
# Entradas: frequency(float), amplitude(float), t(array), waveType(int)
# Salida: wave(ndarray), representa la señal limpia

def waveGenerator(frequency, amplitude, t, waveType):
  # Si la variable waveType es 1, se genera una onda senoidal
  if waveType == 1:
    wave = amplitude * np.sin(2*np.pi*frequency*t)

  # Si la variable waveType es 1, se genera una onda diente de sierra
  elif waveType == 2:
    wave = amplitude * signal.sawtooth(2*np.pi*frequency* t)

  # Si la variable waveType es 1, se genera una onda cuadrada
  elif waveType == 3:
    wave = amplitude * signal.square(2*np.pi*frequency*t)

  # Retorna la onda limpia
  return wave
```

# Definición de las función generadora de señal con ruido

A continuación, se va a generar una función que toma cualquier función y le agrega ruido dependiendo del nivel de ruido deseado por el usuario. En este caso el ruido que se genera es de tipo Gaussiano utilizando la librería de numpy con la siguiente estructura: np.random.normal(media , desviación estándar, len(t)). De acá, el parámetro que define el nivel de ruido de la señal es la desviación estándar, si este valor es mayor, significa que la señal va a ser más ruidosa.

```python
# Función que genera la onda ruidosa con base en la señal limpia y al valor de
# la variable global de array de tiempo y el nivel de ruido deseado (noiseLevel).
# Entradas: wave(ndarray), t(array), noiseLevel(int)
# Salida: noisyWave(ndarray), representa la señal ruidosa

def noiseGenerator(wave, t, noiseLevel):
  # Si la variable noiseLvel es 1, se genera una onda con un nivel de ruido bajo
  if noiseLevel == 1:
    # En esta línea se genera una señal ruidosa el primer parámetro es el
    # promedio, el segundo es la desviación estándar (este es el parámetro que
    # indica el nivel de ruido de la onda) y el tercero es el tamaño del time array
    noise = np.random.normal(0 , 0.01, len(t))

  # Si la variable noiseLvel es 2, se genera una onda con un nivel de ruido medio
  # incrementando el valor de la desviación estándar
  elif noiseLevel == 2:
    noise = np.random.normal(0 , 0.05, len(t))

  # Si la variable noiseLvel es 3, se genera una onda con un nivel de ruido alto
  # incrementando el valor de la desviación estándar
  elif noiseLevel == 3:
    noise = np.random.normal(0 , 0.1, len(t))

  # Se genera una señal que está compuesta de la señal original más el ruido
  noisyWave = wave + noise
  # Se retorna la señal ruidosa
  return noisyWave
```

# Programa principal

En esta parte del código se solicita al usuario el tipo de señal que desea y el nivel de ruido. Luego se llaman a las funciones de generar señal y generar ruido y finalmente se grafican los resultados.

```python
# PRINCIPAL

# Entradas solicitadas al usuario
# waveType: Es de tipo entero y representa el tipo de señal deseada según lo indicado
# noiseLevel: Es de tipo entero y representa el nivel de ruido deseado según lo indicado
waveType = int(input("Seleccione el tipo de onda que desea: Senoidal (1), Diente de sierra (2) o Cuadrada (3): "))
noiseLevel = int(input("Seleccione el nivel de ruido que desea en la señal: Bajo (1), Medio (2) o Alto (3): "))

# Se llama a la función que genera la señal limpia y se le ingresan los siguientes parámetros:
# frecuencia, amplitud, array de tiempo (definidos al inicio) y el tipo de onda deseada
wave = waveGenerator(frequency, amplitude, t, waveType)

# Se llama a la función que genera la señal con ruido y se le ingresan los siguientes parámetros:
# la señal limpia generada, array de tiempo (definido al inicio) y el nivel de ruido deseado
noisywave = noiseGenerator(wave, t, noiseLevel)

# Se grafican los resultados con matplot
plt.plot(t, wave)
# Título del gráfico
plt.title("Onda limpia")
# Etiqueta del eje X
plt.xlabel("Voltage (V)")
# Etiqueta del eje Y
plt.ylabel("Amplitude ")
plt.show()

plt.plot(t, noisywave)
# Título del gráfico
plt.title("Onda con ruido")
# Etiqueta del eje X
plt.xlabel("Voltage (V)")
# Etiqueta del eje Y
plt.ylabel("Amplitude ")
plt.show()
```

En esta parte se va a calcular el porcentaje de distorsión armónica de un sistema. Para esto se le solicita al usuario la cantidad de armónicos que tiene el sistema y luego cada uno de sus valores. El primer valor que ingrese el usuario se sabe que corresponde al armónico de la frecuencia fundamental y por ello se almacena en una variable aparte. Mientras que los demás se almacenan en un array. Posteriormente, con ayuda de la librería numpy se obtiene la norma de todos los elementos del array mencionado y por último solo se aplica la ecuación de la distorsión armónica.

```python
# Entadas dadas por el usuario
# Número de armónicos totales
nArmonics = int(input("Ingrese el número de armónicos: "))
# Se crea un array para almacenar los armónicos no fundamentales
armonics = []

for n in range(nArmonics):
  #Se verifica si es el armónico fundamental y se almacena en una variable aparte
  if n == 0:
    aff = float(input("Ingrese el valor V_RMS del armónico "+ str(n+1)+ ": "))

  # Se almacenan en un array los demás armónicos
  else:
    armonics.append(float(input("Ingrese el valor V_RMS del armónico "+ str(n+1)+ ": ")))

# Se obtiene el módulo o magnitud de los armónicos no fundamentales
magnitude = np.linalg.norm(armonics)

# Se calcula el porcentaje de distorsión armónica y se muesta en pantalla
THDP = (magnitude / aff) * 100
print ("El porcentaje de distorsión armónica es: ", THDP)
```

#Solución Parte 3
En esta parte se tendrá el código correspondiente a los cálculos necesarios para el caso de un amplificador no ideal, como: S/N a la salida en dB, S/N a la entrada en dB, factor de ruido y cifra de ruido. Esto luego de solicitarle los datos requeridos al usuario.

#Librerías y función para realizar los cálculos requeridos
En este bloque del código se importan las librerías necesarias y se crea la función correspondiente a los cálculos requeridos para este enunciado.

```python
# Importar librerías
import numpy as np

# Función de amplificador no ideal
def amplificador_no_ideal(si, ni, ap, nd):
    """
    Calcula las variables asociadas al ruido en un amplificador no ideal.

    Parámetros:
    - Si: Potencia de señal de entrada en Watts.
    - Ni: Potencia de ruido de entrada en Watts.
    - ap: Ganancia de potencia del amplificador.
    - nd: Ruido interno del amplificador.

    Retorna:
    - sn_out_dB: Relación señal a ruido en la salida (S/N a la salida en dB).
    - sn_in_dB: Relación señal a ruido en la entrada (S/N a la entrada en dB).
    - f: Factor de ruido del amplificador.
    - nf: Cifra de ruido del amplificador.
    """
    # Calcular el S/N a la salida
    sn_out = (si * ap) / (ap * ni + nd)

    # Calcular el S/N a la entrada
    sn_in = si / ni

    # Calcular el factor de ruido
    f = sn_in/sn_out

    # Calcular la cifra de ruido
    nf = 10 * np.log10(f)

    # Convertir a dB
    sn_out_dB = 10 * np.log10(sn_out)
    sn_in_dB = 10 * np.log10(sn_in)


    # Devolver los resultados
    return sn_out_dB, sn_in_dB, f, nf
```

#Programa principal
Se realiza una función encargada de mostrar los resultados finales que se obtuvieron, así como la solicitud de los datos requeridos por el usuario. Cabe destacar que el programa acepta notación científica.


```python
def mostrar_resultados(sn_out_dB, sn_in_dB, f, nf):
    """
    Muestra los resultados del cálculo de variables asociadas al ruido en un amplificador no ideal.

    Parámetros:
    - sn_out_dB: Relación señal a ruido en la salida (S/N a la salida en dB).
    - sn_in_dB: Relación señal a ruido en la entrada (S/N a la entrada en dB).
    - f: Factor de ruido del amplificador.
    - nf: Cifra de ruido del amplificador en dB.
    """
    print("\nResultados:")
    print("S/N a la salida en dB:", "{0:.2f}".format(sn_out_dB))
    print("S/N a la entrada en dB:", "{0:.2f}".format(sn_in_dB))
    print("Factor de ruido:", f)
    print("Cifra de ruido en dB:", "{0:.2f}".format(nf))

# Inputs del usuario
si = float(input("Ingrese la potencia de señal de entrada en Watts (Si): "))
ni = float(input("Ingrese la potencia de ruido de entrada en Watts (Ni): "))
ap = float(input("Ingrese la ganancia de potencia (Ap): "))
nd = float(input("Ingrese el ruido interno (Nd): "))

# Llamada a la función amplificador_no_ideal con los datos ingresados por el usuario
sn_out_dB, sn_in_dB, f, nf = amplificador_no_ideal(si, ni, ap, nd)

# Mostrar resultados
mostrar_resultados(sn_out_dB, sn_in_dB, f, nf)
```
